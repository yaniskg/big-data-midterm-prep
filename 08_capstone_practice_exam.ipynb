## Part II — Concept checks (20 pts)

### C1. Logistic regression interpretation
A fitted model gives $\hat\beta_0 = -1.2,\ \hat\beta_1 = 0.5$, with one predictor. Write (a) the expression for the log-odds at $x_1$, and (b) interpret $\hat\beta_1$.

---
**Answer.**
(a) $\log\dfrac{\hat p}{1-\hat p} = -1.2 + 0.5\,x_1$.
(b) A one-unit ↑ in $x_1$ is associated with a **0.5 unit increase in the log-odds**, equivalently the odds are multiplied by $e^{0.5}\approx 1.65$, i.e. odds rise by ~65%.

| Actual 1 | 40 | 60 |

Compute accuracy, sensitivity, specificity.

### C3. Lasso
Consider the objective $\sum_i(y_i - x_i'\beta)^2 + \lambda\sum_j|\beta_j|.$

(a) Name this method.

(b) What happens as $\lambda \to 0$?

(c) Why is standardization important?

---
**Answer.** (a) Lasso. (b) As $\lambda\to 0$ the penalty vanishes, so the solution approaches OLS. (c) Because the penalty treats all coefficients equally regardless of the units of $x$, features measured on large scales will be penalized more lightly and retained preferentially. Standardizing each feature to zero mean and unit variance removes this artifact. (notebook 04)

### C4. Decision threshold
A binary classifier outputs probability $p$ that $Y=1$. Profit if the action is taken and $Y=1$: +200. Cost if the action is taken and $Y=0$: −100. Cost of no action: 0.

(a) Write the expected profit of taking the action as a function of $p$.

(b) Derive the threshold $p^\star$ where we should take action.

---
**Answer.** (a) $\mathbb E[\text{act}\mid p] = 200 p - 100(1-p) = 300 p - 100$. (b) Act iff expected profit exceeds the 0 of not acting: $300 p - 100 > 0 \Leftrightarrow p > 1/3$. $\boxed{p^\star = 1/3 \approx 0.333.}$ (notebook 06)

### C5. KNN
(a) Why can KNN produce non-linear decision boundaries?

(b) Why does feature scaling matter?

---
**Answer.** (a) KNN's decision rule is local and fully non-parametric — the predicted class for any new point depends on the labels of its $K$ nearest *training* neighbours, with no global linear constraint. The boundary is a union of Voronoi cells and can take arbitrary shape. (b) Euclidean distance is scale-sensitive. Without standardization, a feature with a larger unit (e.g., dollars) will dominate the distance computation, so neighbourhoods are determined almost entirely by that one feature. Scaling each feature to unit variance gives all coordinates equal influence. (notebook 07)

## Part III — Short answer (30 pts)

### S1. Probit GLM (15 pts) — recreated from Midterm Practice 1.1

Suppose $Y_i \in \{0, 1\}$ is modeled as $P(Y_i = 1 \mid X_i) = \Phi(X_i'\beta)$ where $\Phi$ is the standard-normal CDF. The sample is iid.

**1.** Write the likelihood $L(\beta; Y, X)$ and log-likelihood $\ell(\beta; Y, X)$.

**2.** Let $\phi(\cdot) = \Phi'(\cdot)$. Show that
$$S(\beta) \equiv \frac{\partial \ell}{\partial \beta} = \sum_{i=1}^n \frac{\phi(X_i'\beta)(Y_i - \Phi(X_i'\beta))}{\Phi(X_i'\beta)(1-\Phi(X_i'\beta))}\,X_i.$$

**3.** What advantage might the probit link have over the logit link?

---
**Answer.**

**Part 1.** With $p_i = \Phi(X_i'\beta)$, each observation contributes $p_i^{Y_i}(1-p_i)^{1-Y_i}$ and, by independence,
$$L(\beta) = \prod_{i=1}^n \Phi(X_i'\beta)^{Y_i}\,\big(1-\Phi(X_i'\beta)\big)^{1-Y_i}.$$
$$\ell(\beta) = \sum_{i=1}^n\Big[Y_i\log\Phi(X_i'\beta) + (1-Y_i)\log(1-\Phi(X_i'\beta))\Big].$$

**Part 2.** Differentiate term by term. Chain rule gives $\partial\Phi(X_i'\beta)/\partial\beta = \phi(X_i'\beta)\,X_i$, so
$$\frac{\partial \log \Phi(X_i'\beta)}{\partial \beta} = \frac{\phi(X_i'\beta)}{\Phi(X_i'\beta)}\,X_i,\qquad \frac{\partial \log (1-\Phi(X_i'\beta))}{\partial \beta} = -\frac{\phi(X_i'\beta)}{1-\Phi(X_i'\beta)}\,X_i.$$
Thus
$$S(\beta) = \sum_{i=1}^n\!\left[\frac{Y_i\phi(X_i'\beta)}{\Phi(X_i'\beta)} - \frac{(1-Y_i)\phi(X_i'\beta)}{1-\Phi(X_i'\beta)}\right] X_i.$$
Putting over a common denominator $\Phi(1-\Phi)$ (dropping $X_i'\beta$ arguments for brevity):
$$S(\beta) = \sum_{i=1}^n \frac{Y_i\phi(1-\Phi) - (1-Y_i)\phi\Phi}{\Phi(1-\Phi)}\,X_i = \sum_{i=1}^n \frac{\phi(Y_i - \Phi)}{\Phi(1-\Phi)}\,X_i.\ \blacksquare$$

**Part 3.** The probit link corresponds to an assumption of **Gaussian** latent errors ($Y^\star = X'\beta + \varepsilon$, $\varepsilon\sim\mathcal N(0,1)$), whereas logit corresponds to Logistic(0,1) errors. If you believe unobserved drivers of the outcome are sums of many small independent shocks (CLT), Gaussian noise is the natural choice — making probit the preferable GLM. Probit also has slightly lighter tails than logit, which matters when modeling rare-event contexts where heavy tails would be unrealistic. In practice the two models produce nearly identical predictions; the main reason to choose probit is compatibility with other Gaussian-based models (e.g., Heckman selection, bivariate probit).

### S2. Business decision problem (15 pts) — fresh scenario

An airline has built a logistic-regression model that outputs $\hat p_i$, the probability that passenger $i$ **misses their connection** given historical flight data.

Actions and payoffs (per passenger):

- **Act** = automatically rebook the passenger onto a later connection. Cost: \$40 (rebooking ops overhead).
  - If the passenger *would have* missed the connection ($Y=1$): rebooking saves \$300 of future goodwill/compensation → **net +\$260**.
  - If the passenger *would have* made the connection ($Y=0$): rebooking caused a \$40 cost and \$60 of passenger inconvenience → **net −\$100**.
- **Don't act** = do nothing. Payoff 0 either way.

**(a)** Write the action-cost matrix.

**(b)** Derive the threshold $t^\star$ on $\hat p$ above which the airline should rebook.

**(c)** Compared to the default $t = 0.5$, does using $t^\star$ increase or decrease the classifier's sensitivity for detecting missed connections? Its specificity? Justify.

**(d)** Suppose the base rate of missed connections is 10%. Using $t = 0.5$ on a well-fit logistic model, would you expect sensitivity or specificity to be higher? Why?

---
**Answer.**

**(a) Action-cost matrix.**

| | True $Y=1$ (misses) | True $Y=0$ (doesn't miss) |
|---|---|---|
| Act (rebook) | +260 | −100 |
| Don't act | 0 | 0 |

**(b) Derive $t^\star$.**
$$\mathbb E[\text{act}\mid p] = 260\,p - 100(1-p) = 360 p - 100.$$
Rebook iff this exceeds the zero payoff of not acting:
$$360\,p - 100 > 0 \Leftrightarrow p > \frac{100}{360} = \frac{5}{18} \approx 0.278.$$
$$\boxed{t^\star \approx 0.278.}$$

Equivalently, $t^\star = C_{FP}/(C_{FP}+B_{TP}) = 100/(100 + 260) = 100/360$, matching the general formula.

**(c)** Moving the threshold from 0.5 to 0.28 **lowers** it, so we rebook more passengers — capturing more of the true missed-connection cases. **Sensitivity goes up** (more true positives among the actual positives), and **specificity goes down** (we also rebook more passengers who would have made it, so the false-positive rate grows). This is the standard trade-off captured by moving along the ROC curve toward the upper-right corner.

**(d)** With base rate 10%, most passengers have $\hat p$ well below 0.5 — the classifier almost always predicts "no rebook". Specificity (correctly identifying the many true negatives) will be very high, while sensitivity (catching the few positives) will be low. This is identical to the behaviour Assignment 4 Q6 observed with Bud Light base rate ~30%: at $t=0.5$ the classifier was far more specific than sensitive.
