## § 1 — Foundations

**Bayes.**&nbsp;&nbsp;&nbsp;$\displaystyle p(\theta\mid d)=\frac{p(d\mid\theta)\,p(\theta)}{p(d)},\qquad \arg\max_\theta p(\theta\mid d)=\arg\max_\theta\,p(d\mid\theta)\,p(\theta).$

**iid joint.**&nbsp;&nbsp;&nbsp;$\displaystyle p(y_1,\ldots,y_n\mid X,\beta)=\prod_{i=1}^n p(y_i\mid x_i,\beta).$

**Three densities.**

| model | $p(y\mid\mu)$ |
|---|---|
| Normal $(\mu,\sigma^2)$ | $\dfrac{1}{\sigma\sqrt{2\pi}}\exp\!\big(-\tfrac{(y-\mu)^2}{2\sigma^2}\big)$ |
| Bernoulli$(p)$ | $p^y(1-p)^{1-y},\ y\in\{0,1\}$ |
| Laplace$(0,b)$ | $\dfrac{1}{2b}\exp(-\lvert y\rvert/b)$ |

**Likelihood chain.**&nbsp;&nbsp;&nbsp;$L(\beta)=p(\mathbf y\mid X,\beta),\ \ \ \ell=\log L,\ \ \ \text{Dev}=-2\ell.$

$$\arg\max L=\arg\max \ell=\arg\min\text{Dev}.$$

*Argmax is preserved by $\log$ and by multiplying/adding constants that don't depend on $\beta$.*

**Paradigms.** Supervised (features + labels), Unsupervised (features only), Reinforcement (rewards). Midterm = supervised only.

## § 2 — Linear Regression

**Model.**&nbsp;&nbsp;$y_i=x_i'\beta+\varepsilon_i,\ \varepsilon_i\overset{iid}{\sim}\mathcal N(0,\sigma^2)$.  
$\Rightarrow$ **MLE = OLS**:&nbsp;&nbsp;$\hat\beta=\arg\min_\beta \text{RSS}(\beta)=\sum(y_i-x_i'\beta)^2.$

**Fit metrics.**

$$\text{RSS}=\sum(\hat y_i-y_i)^2,\quad \text{TSS}=\sum(y_i-\bar y)^2,\quad \text{MSE}=\text{RSS}/n,\quad R^2=1-\text{RSS}/\text{TSS}.$$

**Overfit signature:** train $R^2\gg$ test $R^2$.  
**Underfit signature:** both low.

### Coefficient interpretation — the four cases
| LHS / RHS | meaning of $\beta$ |
|---|---|
| $y\sim x$ | 1-unit $\Delta x \Rightarrow$ $\beta$ units $\Delta y$ |
| $\log y\sim x$ | 1-unit $\Delta x \Rightarrow$ $\approx\!100\beta\%$ $\Delta y$ |
| $y\sim \log x$ | $1\%$ $\Delta x \Rightarrow$ $\beta/100$ units $\Delta y$ |
| $\log y\sim \log x$ | $1\%$ $\Delta x \Rightarrow$ $\beta\%$ $\Delta y$ (**elasticity**) |

**Mnemonic:** log on the side → percent on that side.

### Categorical & interaction mechanics
- `drop_first=True` $\to$ one level is the **reference**; intercept absorbs it.
- Categorical $\hat\beta_{\text{level}}=$ "level − reference" (at the reference value of other $x$).
- Interaction $x_1\cdot x_2$: slope on $x_1$ becomes $\beta_1+\beta_{12}\,x_2$. **Always write out both simple slopes.**
- Polynomial / log transforms keep the model *linear in $\beta$*.

### Omitted-Variable Bias
Short vs. long regression: if true $y=\beta_1 x_1+\beta_2 x_2+\varepsilon$ but we fit $y\sim x_1$ only,
$$E[\hat\beta_1^{\text{short}}]-\beta_1\;=\;\beta_2\cdot\text{Cov}(x_1,x_2)/\text{Var}(x_1).$$
$$\boxed{\ \text{sign(bias)}\;=\;\text{sign}(\beta_2)\,\cdot\,\text{sign}(\text{corr}(x_1,x_2))\ }$$

*Causal coefficients* need **all** variables correlated with both $x^\star$ and $y$ controlled — predictive ≠ causal.

### Key identities used in derivations
$$\frac{\partial}{\partial\beta}\tfrac12\sum(y_i-x_i'\beta)^2\;=\;-\sum(y_i-x_i'\beta)x_i.$$

## § 3 — Logistic / GLM

**Setup.**&nbsp;&nbsp;$Y_i\mid X_i\sim\text{Bernoulli}(p_i),\ p_i=f(x_i'\beta),\ f\uparrow:\mathbb R\to(0,1).$

| link | $f(z)$ | $f'(z)$ |
|---|---|---|
| **Logit** (sigmoid) | $\sigma(z)=\dfrac{e^z}{1+e^z}=\dfrac{1}{1+e^{-z}}$ | $\sigma(z)(1-\sigma(z))$ |
| **Probit** | $\Phi(z)$ (standard-normal CDF) | $\phi(z)$ (std-normal pdf) |

**Log-likelihood.**&nbsp;&nbsp;$\ell(\beta)=\sum_i[y_i\log p_i+(1-y_i)\log(1-p_i)].$

**Binomial deviance.**&nbsp;&nbsp;$\text{Dev}(\beta)=-2\ell(\beta).$

### Log-odds / odds-ratio (THE exam gotcha)
$$\log\frac{p}{1-p}=x'\beta \Longrightarrow \text{odds ratio for }\Delta x_j=1:\ \ \exp(\beta_j).$$
$\beta_j$ is **NOT** a change in probability. 1-unit ↑ $x_j$ ⇒ log-odds ↑ by $\beta_j$ ⇒ odds ×$e^{\beta_j}$.

### Logistic score (memorize)
$$\boxed{\;\frac{\partial\ell}{\partial\beta}=\sum_i(y_i-p_i)\,x_i\;}$$
derived via $\partial\log p_i/\partial\beta=(1-p_i)x_i$, $\partial\log(1-p_i)/\partial\beta=-p_i x_i$.

### Probit score (Midterm Practice 1.1)
$$S(\beta)=\sum_i\frac{\phi(x_i'\beta)(y_i-\Phi(x_i'\beta))}{\Phi(x_i'\beta)(1-\Phi(x_i'\beta))}\,x_i.$$

### Multinomial (softmax)
$$P(Y=k\mid x)=\frac{e^{x'\beta_k}}{\sum_j e^{x'\beta_j}},\qquad \hat y=\arg\max_k x'\beta_k.$$

### Gradient descent (why logistic needs it)
$$\beta_{t+1}=\beta_t-\eta\,\nabla\text{Dev}(\beta_t);\ \ \ \text{stop when }|\beta_{t+1}-\beta_t|<\text{tol}.$$
Logistic has no closed form (score equation is nonlinear); linear does ($X'X\hat\beta=X'y$).

### Logit vs probit — when to prefer
Probit ⇔ latent $Y^\star=x'\beta+\varepsilon$ with $\varepsilon\sim\mathcal N(0,1)$. Use probit when you want Gaussian-error assumption (CLT aggregation, plug into Heckman selection models, etc.). Logit has the closed-form log-odds interpretation and is slightly faster.

## § 4 — Regularization

| method | objective | $\lambda\to 0$ | $\lambda\to\infty$ | selects? |
|---|---|---|---|---|
| **Lasso (L1)** | $\text{RSS}+\lambda\sum\lvert\beta_j\rvert$ | OLS | all $\hat\beta=0$ | **yes** (cusps at 0) |
| **Ridge (L2)** | $\text{RSS}+\lambda\sum\beta_j^2$ | OLS | all $\hat\beta\to0$ | no (smooth) |

$\lambda\uparrow$: bias↑, variance↓, training RSS↑, test RSS U-shaped.

### Always standardize features before L1/L2
Penalty is unit-sensitive. Without scaling, large-unit features keep larger coefs and are preferentially retained regardless of signal. Fix: $\tilde x_j=(x_j-\bar x_j)/\hat\sigma_{x_j}$.

### Shrinkage bias
Even selected features have lasso coefficients biased toward 0. **"Relaxed lasso"**: lasso for selection → OLS refit on the chosen subset.

### Bayesian MAP correspondence (memorize derivation)

| prior on $\beta_j$ iid | MAP | $\lambda$ |
|---|---|---|
| Laplace$(0,b)$ | **Lasso** | $2\sigma^2/b$ |
| Normal$(0,\tau^2)$ | **Ridge** | $\sigma^2/\tau^2$ |

**4-line MAP derivation template (fits any prior).**
1. $\log p(\beta\mid d)=\log L(\beta)+\log p(\beta)+\text{const}.$
2. Substitute Gaussian likelihood $\Rightarrow -\tfrac{1}{2\sigma^2}\text{RSS}$.
3. Substitute prior: Laplace $\Rightarrow -\tfrac{1}{b}\|\beta\|_1$; Normal $\Rightarrow -\tfrac{1}{2\tau^2}\|\beta\|_2^2$.
4. Multiply by $-2\sigma^2$ (flip argmax → argmin), drop constants.

## § 5 — Model Selection

### Forward stepwise
Greedy. Start empty; at each step add the feature giving the lowest training RSS. Fits $O(kp)$ OLS regressions for $k$ features from $p$. Uses **unshrunken** coefficients. Does *not* guarantee global best subset.

### Information criteria
$$\text{AIC}=-2\ell+2k,\qquad \text{BIC}=-2\ell+k\log n.$$
Lower wins. BIC penalizes more than AIC whenever $n>e^2\approx 7.4$.

### $K$-fold CV arithmetic
| quantity | formula | $n=1000,K=5$ |
|---|---|---|
| # fits | $K$ | 5 |
| rows per fit | $n(K-1)/K$ | 800 |
| rows per val fold | $n/K$ | 200 |

*If also tuning $\lambda$ over $G$ grid points: total $K\!\cdot\!G$ fits.*

### CV-min vs 1-SE rule
- **CV-min:** pick $\lambda$ minimizing mean CV loss.
- **1-SE:** among $\lambda$ whose CV loss $\le$ min + 1 SE, pick the **largest** (simplest).

### `TimeSeriesSplit`
Expanding-window splits. Fold $k$ trains on the earliest $\tfrac{k}{K+1}$, validates on the *next* contiguous block. **No look-ahead leakage.** Use for any time-series data.

### Why AIC/BIC fail on non-iid data
Both use raw $n$ as the complexity weight. Autocorrelation $\Rightarrow$ effective sample size $\ll n \Rightarrow$ penalty too small $\Rightarrow$ models over-selected. CV measures generalization empirically, so it degrades more gracefully.

## § 6 — Classification Evaluation

### Confusion matrix (memorize the orientation!)

| | **Predicted 0** | **Predicted 1** |
|---|---|---|
| **Actual 0** | TN | FP |
| **Actual 1** | FN | TP |

$n_P=\text{TP}+\text{FN}$ (positives), $n_N=\text{TN}+\text{FP}$ (negatives).

### Six rates
| name | formula | denominator |
|---|---|---|
| Accuracy | $(TP+TN)/n$ | everyone |
| **Sensitivity = TPR = Recall** | $TP/n_P$ | within positives |
| **Specificity = TNR** | $TN/n_N$ | within negatives |
| FPR | $FP/n_N = 1-$TNR | within negatives |
| FNR | $FN/n_P = 1-$TPR | within positives |
| Precision (PPV) | $TP/(TP+FP)$ | within "predict 1"s |

**Direction of threshold moves:**
$$t\downarrow \Rightarrow \text{sensitivity}\uparrow,\ \text{specificity}\downarrow$$
$$t\uparrow \Rightarrow \text{sensitivity}\downarrow,\ \text{specificity}\uparrow$$

### ROC construction
Sort observations by $\hat p_i$ ↓. Sweep $t$ from 1 → 0. At each $t$, classify $\hat y=\mathbb 1(\hat p>t)$ and plot point $(FPR,TPR)$.

$$\text{AUC}=\Pr(\hat p_{y=1}>\hat p_{y=0})\in[0,1].$$
0.5 = chance; 1 = perfect ranker. **Threshold-free.**

### Decision-theoretic threshold (THE business problem)
Build action-payoff matrix:

| | True $Y=1$ | True $Y=0$ |
|---|---|---|
| Act | $+B_{TP}$ | $-C_{FP}$ |
| Don't | 0 | 0 |

$\mathbb E[\text{act}\mid p]=B_{TP}\,p-C_{FP}(1-p)$. Act iff > 0:
$$\boxed{\ t^\star=\frac{C_{FP}}{C_{FP}+B_{TP}}\ }$$

**Variants** — if don't-act has non-zero costs, always compute $\mathbb E[\text{act}]-\mathbb E[\text{don't}]$, set to 0, solve for $p$. Example (tutoring, slide 16): $-30p+90(1-p) > -25(1-p)\Rightarrow p<23/29\approx0.79$.

### Youden's J
$J(t)=\text{TPR}(t)-\text{FPR}(t)$. $t^\star$ maximizing $J$ = point on ROC furthest above the 45° line.

### Base-rate effect
If positive base rate $\pi\ll 0.5$ and $t=0.5$: model rarely predicts 1 ⇒ high specificity, low sensitivity (this is *mechanical*, not bad training). Lower $t$ (or use cost formula) to rebalance.

## § 7 — KNN

**Rule.**&nbsp;&nbsp;$\hat y_{\text{new}}=\arg\max_j\ \tfrac1K\sum_{k=1}^K\mathbb 1\{y_{i_k}=j\}$ (majority vote of the $K$ closest training points). Regression: average instead of mode.

### Distance choice
| data | distance |
|---|---|
| continuous (scaled!) | **Euclidean** $\sqrt{\sum(u_j-v_j)^2}$ |
| categorical / 0-1 dummies | **Hamming** $\sum\mathbb 1\{u_j\neq v_j\}$ |

### Bias–variance in $K$
| $K$ | bias | variance | train acc | boundary |
|---|---|---|---|---|
| 1 | 0 | max | 100% (if rows unique) | jagged |
| small | low | high | high | wiggly |
| mid | moderate | moderate | moderate | smooth |
| $n$ | max (modal class) | 0 | base rate | constant |

**$K$ is the complexity knob; $K\!\downarrow$ = more overfit (opposite direction to $\lambda$ in lasso).**

### Key rules
- **Scale** continuous features before Euclidean — variance 100 dominates variance 1.
- Hamming on 0/1 dummies is already scale-free.
- Prediction cost $O(np)$ per query (no training phase).
- Curse of dimensionality: in high $p$, all points become ≈ equidistant.
- **Non-linear boundary** = union of Voronoi cells; captures patterns logistic cannot.

## § 8 — Pros/Cons summary

### If true model is linear
| method | pro | con |
|---|---|---|
| Linear reg | best with small $p$ | overfits with large $p$ |
| Lasso | robust to large $p$ | shrinkage bias; need to tune $\lambda$ |
| KNN | flexible | high variance |

### If true model is unknown
| method | pro | con |
|---|---|---|
| Linear reg | interpretable | overfits if you add non-linear terms |
| Lasso | can safely include many non-linear features | tune $\lambda$; must hand-engineer |
| KNN | non-linear boundary, simple | doesn't scale with $p$; no interpretability |

## § 9 — Derivation Templates (memorize these)

### T1. Gaussian MLE = OLS  *(3 steps)*
1. $p(y_i\mid x_i,\beta)=\tfrac{1}{\sigma\sqrt{2\pi}}\exp(-\tfrac{(y_i-x_i'\beta)^2}{2\sigma^2}).$
2. $\ell(\beta)=-\tfrac n2\log(2\pi\sigma^2)-\tfrac{1}{2\sigma^2}\text{RSS}(\beta).$
3. Drop constants ⇒ $\arg\max\ell=\arg\min\text{RSS}.$

### T2. Logistic score  *(3 steps)*
1. $\ell=\sum[y_i\log p_i+(1-y_i)\log(1-p_i)]$.
2. Use $\sigma'=\sigma(1-\sigma)$: $\ \partial\log p_i/\partial\beta=(1-p_i)x_i,\ \partial\log(1-p_i)/\partial\beta=-p_ix_i$.
3. Combine: $\partial\ell/\partial\beta=\sum(y_i-p_i)\,x_i$.

### T3. Probit score  *(3 steps)*
1. $\ell=\sum[y_i\log\Phi_i+(1-y_i)\log(1-\Phi_i)]$ where $\Phi_i=\Phi(x_i'\beta)$.
2. $\partial\log\Phi_i/\partial\beta=\phi_i/\Phi_i\,x_i,\ \partial\log(1-\Phi_i)/\partial\beta=-\phi_i/(1-\Phi_i)\,x_i.$
3. Common denominator $\Phi_i(1-\Phi_i)$ ⇒ $S(\beta)=\sum\frac{\phi_i(y_i-\Phi_i)}{\Phi_i(1-\Phi_i)}\,x_i$.

### T4. Lasso as MAP (Laplace prior)  *(4 steps)*
1. $\log p(\beta\mid d)=\log L(\beta)+\log p(\beta)+c.$
2. Gaussian $L$: $\log L=-\tfrac{1}{2\sigma^2}\text{RSS}+c'.$
3. Laplace prior: $\log p(\beta)=-\tfrac{1}{b}\sum\lvert\beta_j\rvert+c''.$
4. Multiply by $-2\sigma^2$, drop constants:
$$\hat\beta_{\text{MAP}}=\arg\min\text{RSS}+\tfrac{2\sigma^2}{b}\lVert\beta\rVert_1.$$

### T5. Optimal threshold  *(3 steps)*
1. Write payoff matrix (rows = actions, cols = true states).
2. $\mathbb E[\text{act}\mid p]$, $\mathbb E[\text{don't}\mid p]$ as functions of $p$.
3. Act iff difference > 0, solve linearly for $p$ ⇒ $t^\star$.

### T6. $R^2$ from an output  *(1 step)*
$R^2 = 1-\text{RSS}/\text{TSS}$. Given RSS and $n$, MSE = RSS/$n$.

### T7. Odds ratio from a logistic coefficient  *(1 step)*
Odds $\times$ $e^{\beta_j}$ per unit ↑ in $x_j$.

## § 10 — Exam Traps & Mnemonics

### Wrong answers that look right
| tempting but wrong | actually |
|---|---|
| "$\beta$ in logistic = change in probability" | **change in log-odds** |
| "$\lambda\uparrow$ ⇒ variance↑" | variance ↓ (simpler model) |
| "lasso $\lambda\to 0$ ⇒ zeros out coefs" | $\lambda\to 0$ ⇒ OLS; zeros happen at $\lambda$ large |
| "Training $R^2$ falls as we add features" | Training $R^2$ weakly **rises** |
| "Euclidean KNN on categorical dummies" | Use Hamming for categorical |
| "K=1 has high bias" | K=1 has **low bias, high variance** |
| "BIC picks bigger models than AIC" | BIC picks **smaller** (heavier penalty for $n>7$) |
| "Shuffled CV works fine for time series" | No — use TimeSeriesSplit |

### Sanity-check rules during the exam
- Any "log-odds" question: coefficient is *additive* on the log scale, *multiplicative* ($e^\beta$) on the odds scale.
- Confusion matrix: **read the axes first**. TPR's denominator is the row total for Actual 1.
- Whenever you see "Gaussian errors" → immediately write $L\propto\exp(-\text{RSS}/2\sigma^2)$ and collapse.
- Decision-theoretic threshold: $t^\star = (\text{cost of false alarm})/(\text{cost of false alarm} + \text{benefit of true positive})$.
- Lower threshold = more alarms = more true + more false positives.
- Deviance = $-2\log L$. If they show a deviance, you can retrieve $\ell = -\text{Dev}/2$.

### Common numerical shortcuts
- $e^{0.7}\approx 2$,&nbsp; $e^{1}\approx 2.72$,&nbsp; $e^{-0.7}\approx 0.5$,&nbsp; $e^{0.4}\approx 1.5$.
- $\log(2)\approx 0.69$,&nbsp; $\log(10)\approx 2.30$,&nbsp; $\log(1000)\approx 6.9$.
- $\sigma(0)=0.5$; $\sigma'(0)=0.25$; $\Phi(0)=0.5$; $\phi(0)\approx 0.399$.
- For $n=1000$, $\log n \approx 6.9$ → BIC adds ~$6.9 k$ while AIC adds $2k$.

## § 11 — One-line summaries (when you have 30 seconds per problem)

- **Overfit?** Training $R^2$ much higher than test $R^2$.
- **Which is non-parametric?** KNN.
- **λ↑ in lasso?** More coefficients zero; bias ↑, variance ↓.
- **K↓ in KNN?** More overfit; bias ↓, variance ↑.
- **β in logistic?** Change in log-odds. Odds × $e^\beta$.
- **Linear regression minimizes?** RSS (equivalently maximizes Gaussian likelihood).
- **Deviance?** $-2\log L$; minimized during estimation of any GLM.
- **Log-log coefficient?** Elasticity (% per %).
- **OVB sign?** sign($\beta_{\text{omitted}}$) × sign(corr($x_{\text{included}}$, $x_{\text{omitted}}$)).
- **1-SE rule picks?** Largest λ with CV ≤ min + 1 SE — most parsimonious nearly-best model.
- **t\* (send-a-mailer)?** $C_{FP}/(C_{FP}+B_{TP})$.
- **Lower t?** Sens↑, Spec↓.
- **KNN distance for dummies?** Hamming.
- **KNN needs scaling?** Yes, for Euclidean.
- **Why probit?** Gaussian latent error (CLT story / Heckman compatibility).
- **Why gradient descent for logistic but not linear?** Score equation nonlinear in $\beta$; no closed form.
- **K-fold CV fits?** $K$, each using $(K-1)n/K$ rows.
- **TimeSeriesSplit?** Expanding training window, validation strictly after training. No leakage.
- **Why AIC/BIC fail on autocorrelated data?** They use raw $n$; effective $n$ is smaller → under-penalize complexity.
- **Laplace prior ↔ ?** Lasso MAP. Gaussian prior ↔ Ridge.
- **AUC = 0.5?** Random. AUC = 1? Perfect ranker.
